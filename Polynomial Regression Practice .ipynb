```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
df
```

```python
sns.pairplot(df)
```

```python
df.plot(x ='LSTAT', y ='MEDV', style ='o')
plt.title('Housing Dataset')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.show()
```

```python
X = df[['LSTAT']]
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression    #Fitting linear regression to dataset
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
def viz_linear():                                    #visualization of linear regression                                   
    plt.scatter(X, y, color='blue')
    plt.plot(X, regressor.predict(X), color='red')
    plt.title('Linear Regression')
    plt.xlabel('LSTAT')
    plt.ylabel('MEDV')
    plt.show()
    return
viz_linear()
```

```python
from sklearn.preprocessing import PolynomialFeatures      #Fitting polynomial regression to datatset
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)
```

```python
def viz_pol():                                            #Visualization after fitting polynomial regression 
    plt.scatter(X, y, color='blue')
    plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='red')
    plt.title('Linear Regression')
    plt.xlabel('LSTAT')
    plt.ylabel('MEDV')
    plt.show()
    return
viz_pol()
```

### Comparing Polynomial Degrees 

```python
poly_reg = PolynomialFeatures(degree=1)
X_poly = poly_reg.fit_transform(X)
pol_reg.fit(X_poly, y)
ypred1 = pol_reg.predict(X_poly)
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
pol_reg.fit(X_poly, y)
ypred2 = pol_reg.predict(X_poly)
```

```python
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)
pol_reg.fit(X_poly, y)
ypred3 = pol_reg.predict(X_poly)
```

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
pol_reg.fit(X_poly, y)
ypred4 = pol_reg.predict(X_poly)
```

```python
plt.scatter(X,y,label="data points")
plt.plot(X, ypred1, color='red', label="Degree 1")
plt.plot(X, ypred2, color='magenta', label="Degree 2")
plt.plot(X, ypred3, color='blue', label="Degree 3")
plt.plot(X, ypred4, color='green', label="Degree 4")
plt.legend()
```
